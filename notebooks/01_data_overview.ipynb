````python
from pathlib import Path
import pandas as pd
import numpy as np

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = ROOT / "data" / "processed"

csv_files = sorted(DATA_DIR.glob("*.csv"))

file_summary = pd.DataFrame({
    "file": [path.name for path in csv_files],
    "size_kb": [round(path.stat().st_size / 1024, 2) for path in csv_files],
})

file_summary
````

````python
training = pd.read_csv(DATA_DIR / "match_training_data.csv")
teams = pd.read_csv(DATA_DIR / "wc2026_teams_model_input.csv")
team_features = pd.read_csv(DATA_DIR / "team_form_elo_latest.csv")

print("training shape:", training.shape)
print("teams shape:", teams.shape)
print("team_features shape:", team_features.shape)
````

````python
# Missing-value check for the main training table
missing = (
    training.isna()
    .sum()
    .reset_index()
    .rename(columns={"index": "column", 0: "missing_count"})
    .sort_values("missing_count", ascending=False)
)

missing.head(25)
````

````python
# Target mapping:
# 0 = Team A loss
# 1 = Draw
# 2 = Team A win

target_counts = (
    training["target_result"]
    .value_counts(normalize=False)
    .sort_index()
    .rename_axis("target_result")
    .reset_index(name="count")
)

target_counts["percentage"] = target_counts["count"] / target_counts["count"].sum()
target_counts
````

````python
# Confirm exactly 48 teams are available for prediction.
teams_per_group = (
    teams.groupby("group_code")["team_name"]
    .count()
    .reset_index(name="team_count")
)

print("Total teams:", teams["team_id"].nunique())
teams_per_group
````
